```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from adspy_shared_utilities import plot_decision_tree
iris_dataset = load_iris()
X = iris_dataset['data']
y = iris_dataset['target']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
dt_clfr = DecisionTreeClassifier()
dt_clfr.fit(X_train,y_train)
print "Training set accuracy of Decision Tree: {:.2f}".format(dt_clfr.score(X_train,y_train))
print "Test set accuracy of Decision Tree: {:.2f}".format(dt_clfr.score(X_test,y_test))
```

Note that, above model is overfitting  
Setting max depth of tree to help avoiding overfitting  

```python
another_dt_clfr = DecisionTreeClassifier(max_depth=3)
another_dt_clfr.fit(X_train,y_train)
print "Training set accuracy of Decision Tree: {:.2f}".format(another_dt_clfr.score(X_train,y_train))
print "Test set accuracy of Decision Tree: {:.2f}".format(another_dt_clfr.score(X_test,y_test))
```
